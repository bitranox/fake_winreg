## 3) Python API: Create and query registry keys

```python
import fake_winreg as winreg

# Create a registry key and set a value
key = winreg.CreateKey(winreg.HKEY_CURRENT_USER, r"Software\TestApp")
winreg.SetValueEx(key, "Version", 0, winreg.REG_SZ, "1.0.0")
winreg.CloseKey(key)

# Read the value back
key = winreg.OpenKey(winreg.HKEY_CURRENT_USER, r"Software\TestApp")
value, reg_type = winreg.QueryValueEx(key, "Version")
print(f"Version = {value!r} (type={reg_type})")
winreg.CloseKey(key)
```

## 4) Enumerate subkeys and values

```python
import fake_winreg as winreg

# Create some subkeys
for name in ("SubA", "SubB"):
    sub = winreg.CreateKey(winreg.HKEY_CURRENT_USER, rf"Software\TestApp\{name}")
    winreg.CloseKey(sub)

# Enumerate subkeys
key = winreg.OpenKey(winreg.HKEY_CURRENT_USER, r"Software\TestApp")
index = 0
while True:
    try:
        subkey_name = winreg.EnumKey(key, index)
        print(f"Subkey[{index}]: {subkey_name}")
        index += 1
    except OSError:
        break
winreg.CloseKey(key)
```
